The performance of the model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from pathlib import Path
from tqdm import tqdm
import os

# === CONFIGURATION ===
# Paths to your model and data
model_path = "/content/drive/MyDrive/tern_project/Eyal/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/terns-detection/weights/best.pt"
test_images_dir = "/content/drive/MyDrive/tern_project/Eyal/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/test/images"
ground_truth_labels_dir = "/content/drive/MyDrive/tern_project/Eyal/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/test/labels"
predictions_save_dir = "/content/drive/MyDrive/tern_project/Eyal/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/test/predictions"  # Temporary predictions folder
class_names = ['Chick', 'Common-Fly', 'Common-Sit', 'Common-Stand', 'Little-Fly', 'Little-Sit', 'Little-Stand', 'Other']  # UPDATE if needed

# Clean/create predictions directory
os.makedirs(predictions_save_dir, exist_ok=True)
for f in Path(predictions_save_dir).glob("*.txt"):
    f.unlink()




```

```python

from ultralytics import YOLO

# === 1. Load Model ===
model = YOLO(model_path)

# === 2. Run Inference on All Test Images ===
image_files = sorted(Path(test_images_dir).glob("*.jpg"))
for img_path in tqdm(image_files, desc="Running YOLO inference"):
    results = model(img_path, conf=0.25, save=False)[0]

    # Save predictions as YOLO-format .txt
    pred_txt = Path(predictions_save_dir) / (img_path.stem + ".txt")
    with open(pred_txt, "w") as f:
        for box in results.boxes.data:
            cls_id = int(box[5].item())
            conf = box[4].item()
            xywh = box[:4].tolist()  # x1, y1, x2, y2
            # Convert to YOLO (x_center, y_center, w, h) format â€” optional for this step
            f.write(f"{cls_id} {conf:.4f}\n")  # Save just class for evaluation
```

```python
# === 3. Evaluation ===
def load_labels(path):
    if not os.path.exists(path):
        return []
    with open(path, 'r') as f:
        return [int(line.strip().split()[0]) for line in f if line.strip()]

all_true = []
all_pred = []

for gt_file in sorted(Path(ground_truth_labels_dir).glob("*.txt")):
    pred_file = Path(predictions_save_dir) / gt_file.name

    gt_labels = load_labels(gt_file)
    pred_labels = load_labels(pred_file)

    # Pad with -1 if prediction/ground truth have unequal lengths
    max_len = max(len(gt_labels), len(pred_labels))
    gt_labels += [-1] * (max_len - len(gt_labels))
    pred_labels += [-1] * (max_len - len(pred_labels))

    all_true.extend(gt_labels)
    all_pred.extend(pred_labels)

# === 4. Compute Metrics ===
accuracy = accuracy_score(all_true, all_pred)
precision_weighted = precision_score(all_true, all_pred, average='weighted', zero_division=0)
recall_weighted = recall_score(all_true, all_pred, average='weighted', zero_division=0)
f1_weighted = f1_score(all_true, all_pred, average='weighted', zero_division=0)

precision_macro = precision_score(all_true, all_pred, average='macro', zero_division=0)
recall_macro = recall_score(all_true, all_pred, average='macro', zero_division=0)
f1_macro = f1_score(all_true, all_pred, average='macro', zero_division=0)

precision_micro = precision_score(all_true, all_pred, average='micro', zero_division=0)
recall_micro = recall_score(all_true, all_pred, average='micro', zero_division=0)
f1_micro = f1_score(all_true, all_pred, average='micro', zero_division=0)

# === 5. Print Results ===
print(f"\nâœ… YOLOv8 Classification Evaluation:\n")
print(f"Accuracy:             {accuracy:.2f}")
print(f"Precision (Weighted): {precision_weighted:.2f}")
print(f"Recall    (Weighted): {recall_weighted:.2f}")
print(f"F1 Score  (Weighted): {f1_weighted:.2f}")
print(f"Precision (Macro):    {precision_macro:.2f}")
print(f"Recall    (Macro):    {recall_macro:.2f}")
print(f"F1 Score  (Macro):    {f1_macro:.2f}")
print(f"Precision (Micro):    {precision_micro:.2f}")
print(f"Recall    (Micro):    {recall_micro:.2f}")
print(f"F1 Score  (Micro):    {f1_micro:.2f}")

# Per-class
print("\nğŸ“‹ Classification Report:")
print(classification_report(all_true, all_pred, target_names=class_names, zero_division=0))
```

Adding augmentation

```python
from ultralytics import YOLO
import os

# Set memory-friendly config
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# Load pretrained model
model = YOLO("yolov8x.pt")

# Train with custom augmentations
model.train(
    data="/content/drive/MyDrive/tern_project/Chicks/terns-project-chick/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/data.yaml",
    epochs=50,
    imgsz=1024,
    batch=4,
    patience=20,
    project="/content/drive/MyDrive/tern_project/Chicks/terns-project-chick/YoloDetector/TrainedVersions/Terns Detection 3.0.v3i.yolov8/",
    name="terns-detection_augmented_x",
    augment=True,           # Enable strong augmentation
    degrees=5.0,            # Small rotations
    translate=0.1,          # Slight shifts
    scale=0.5,              # Zoom in/out
    shear=1.0,              # Small shearing
    perspective=0.0005,     # Keep perspective effect minimal
    flipud=0.0,             # No vertical flips (not logical for chicks)
    fliplr=0.5,             # Horizontal flip
    mosaic=1.0,             # Strong mosaic (helps small objs)
    mixup=0.2,              # Mild MixUp
    hsv_h=0.015,            # Hue jitter
    hsv_s=0.7,              # Saturation jitter
    hsv_v=0.4               # Value jitter
)
```
